```
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
wd = webdriver.Chrome('chromedriver',options=chrome_options)
driver =webdriver.Chrome('chromedriver',options=chrome_options)
```

```
driver.get("https://seoulmusicawards.com/vote.html")
```

```
import time
time.sleep(2)

tabs = driver.find_elements_by_class_name("nav-link")

tab = tabs[0]

tab.click()
```

```
rank = driver.find_elements_by_class_name("rank")

tit = driver.find_elements_by_class_name("tit")

sub = driver.find_elements_by_class_name("sub-tit")

current = driver.find_elements_by_class_name("current")

per = driver.find_elements_by_class_name("per")
```

```
position = [int(i.text) for i in rank[0:50]]
name = [i.text for i in tit[3:53]]
album = [i.text for i in sub[0:50]]
votes = [int(i.text.replace(",", ""))/1000 for i in current[0:50]]
percent = [int(i.text.replace("%", "").replace(".", ""))/100 for i in per[0:50]]
```

### CREATE DATAFRAME

```
import pandas as pd
```

```
data = {"Position": position, "Artist": name, "Album": album, "Votes": votes, "%": percent}
df = pd.DataFrame(data, columns = list(data.keys()))
```

```
df.head()
```

CREATES COLUMN WITH HOW MANY DAYS ARE LEFT UNTIL THE END OF THE POLL

```
from datetime import datetime

now = datetime.now()
d = datetime(2021, 1, 24, 14, 59)

duration = now-d

days = duration.days
seconds = duration.seconds
hours = seconds//3600

time_left = days + (hours/24)
```

```
df["Days Left"] = time_left
df["Time"] = now
```
